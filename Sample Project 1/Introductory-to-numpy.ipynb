```python
# Standard deviation = a measure of how spread out a group of numbers is from the mean
np.std(a2)
```

```python
# Variance = measuer of the average degree to which each number is different to the mean
# Higher varinace = wider range of numbers
# Lower variance = lowder range of numbers
np.var(a2)
```

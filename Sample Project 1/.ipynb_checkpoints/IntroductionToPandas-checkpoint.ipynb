## Dealing with Missing data

```python
car_sales_missing = pd.read_csv("car-sales-extended-missing-data.csv")
car_sales_missing
```

### Either you go re-asignment or inplace 

```python
car_sales_missing['Odometer (KM)'].fillna(car_sales_missing['Odometer (KM)'].mean())
```

```python
car_sales_missing['Odometer (KM)'].mean()
```

```python
car_sales_missing.head(20)
```

```python
car_sales_missing['Odometer (KM)'].fillna(car_sales_missing['Odometer (KM)'].mean(),inplace=True)
```

```python
car_sales_missing.head(20)
```

```python
car_sales_missing_dropped_NaN = car_sales_missing.dropna()
```

```python
car_sales_missing_dropped_NaN.head(20)
```

# Column from python list (As it is from the list, has to be same length)

```python
fuel_economy = [7.5,9.2,5.0,9.6,8.7,4.7,7.6,8.7,3.0,4.5]
car_sales["Fuel per 100KM"] = fuel_economy
car_sales
```

```python
car_sales["Total fuel used (L)"] = car_sales["Odometer (KM)"]/100*car_sales["Fuel per 100KM"]
car_sales
```

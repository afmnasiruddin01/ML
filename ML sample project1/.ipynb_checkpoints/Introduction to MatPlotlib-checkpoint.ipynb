# Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3,4]
y= [11,22,33,44]

# 2. Setup plot
fig, axs = plt.subplots(2,2, sharex=True,figsize=(8,5))

# 3. Plot data
axs[0,1].plot(x,y)

# 4. Customize plot
fig.set(facecolor="lightblue")
fig.suptitle("Sample Title of 2 by 2 subplots",fontweight="bold")

axs[0,1].set_title("Simple Plot")
axs[0,1].set_xlabel("x-axis")
axs[0,1].set_ylabel("y-axis")

line1, =axs[0,0].plot(x,y,label = "Top Left")
line2, =axs[0,1].plot(x,y,label = "Top Right")

fig.legend(
    handles=[line1,line2],
    bbox_to_anchor=(0.3, 1),
    fontsize=10,
    frameon=True
)
fig.savefig("./images/sample-plot.png")
```

### Which one should you use? (pyplot vs matplotlib OO method)
* When plotting something quickly, okay to use the pyplot method.....plot()
* when plotting something more advanced, use OO method.

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50
```

```python
pd.plotting.scatter_matrix(frame=pd.DataFrame(np.random.randn(1000,4)),alpha=0.2)
```

```python
over_50.plot(kind="scatter", 
             x='age',
             y='chol',
             c='target'
            )
```

```python
# OO method mixed with pyplot method
fix, ax =  plt.subplots(figsize=(10,6))
over_50.plot(
    kind="scatter",
    x="age",
    y="chol",
    c="target",
    ax=ax
)
```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# plot the data
scatter = ax.scatter(
    x=over_50["age"],
    y=over_50["chol"],
    c=over_50["target"]
)

# customize the plot
ax.set(title = "Heart Disease and Cholestorol levels",
       xlabel ="Age",
       ylabel="Cholesterol"
      )
ax.legend(*scatter.legend_elements(),title="Target");
# add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle="--")
```

```python
# subplot of chol, age, thalach
fig,(ax0,ax1)= plt.subplots(2,1,figsize=(10,10),sharex=True)
ax0.scatter(
    x=over_50["age"],
    y=over_50["chol"],
    c=over_50["target"]    
)
ax0.set(
    title="Heart Disease and cholesterol levels",
    ylabel="Cholesterol"
)
handles, labels =scatter.legend_elements()
ax0.legend(handles,labels,title="Target")
ax0.axhline(y=over_50["chol"].mean(), linestyle="--")


ax1.scatter(
    x=over_50["age"],
    y=over_50["thalach"],
    c=over_50["target"]    
)
ax1.set(
    title="Heart Disease and thalach levels",
    xlabel="Age",
    ylabel="thalach"
)
handles, labels =scatter.legend_elements()
ax1.legend(*scatter.legend_elements(),title="Target")
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--")

fig.suptitle("The Heart Disease Analysis",fontweight="bold");
```

```python
x=np.random.randn(10,4)
x
```

```python
df=pd.DataFrame(x,columns=["a","b","c","d"])
df
```

```python
ax =df.plot(kind="bar")
ax.set(title = "Random Number Bar Graph from Dataframe",
        xlabel = "Row Number",
       ylabel="Random Number"
      )
ax.legend().set_visible(True)
```

```python
# Set the style
plt.style.use("seaborn-v0_8")

# subplot of chol, age, thalach
fig,(ax0,ax1)= plt.subplots(2,1,figsize=(10,10),sharex=True)
scatter0 = ax0.scatter(
    x=over_50["age"],
    y=over_50["chol"],
    c=over_50["target"],
    cmap="winter" #this change the color schema
)
ax0.set(
    title="Heart Disease and cholesterol levels",
    ylabel="Cholesterol"
)
handles, labels =scatter0.legend_elements(prop="colors",num=2)
ax0.legend(handles,labels,title="Target")
ax0.axhline(y=over_50["chol"].mean(), linestyle="--")


scatter1 = ax1.scatter(
    x=over_50["age"],
    y=over_50["thalach"],
    c=over_50["target"] ,
    cmap="PuBu"   #this change the color schema
)
ax1.set(
    title="Heart Disease and thalach levels",
    xlabel="Age",
    ylabel="thalach"
)
handles, labels =scatter1.legend_elements(prop="colors")
ax1.legend(*scatter1.legend_elements(prop="colors", num=2),title="Target")
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--")

fig.suptitle("The Heart Disease Analysis",fontweight="bold");
```
